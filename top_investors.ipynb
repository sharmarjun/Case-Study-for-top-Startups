```python
''' In this problem we have to find the top 5 investors who have invested maximum number of times '''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("Downloads/startup_funding.csv")
df=data.copy()
dip=df['InvestorsName']
d={}
""" As there are multiple Investors present in single Investor Name so we have to extract all the investor who are involved 
in the funding of a particular startup"""
for ele in dip:
    p=str(ele)
    q=p.split(",")
    for e in q:
        e=e.strip()
        if e in d:
            d[e]+=1
        else:
            d[e]=1

investor=[]
city=[]
count=[]
"""Now we have to find the top 5 investors who involved in the funding of maximum number of startups"""
for i in range(5):
    max=-1
    for ele in d:
        if ele!="":
            if max<d[ele]:
                max=d[ele]
                ct=ele
    count.append(max)
    investor.append(ct)
    del d[ct]
print("Investor  \t      |   No_of_time_invest")
for i in range(5):
    print('{:<22s}|{:>11d}'.format(investor[i],count[i]))
print()
plt.pie(count,labels=investor,autopct="%.2f")
plt.show()
plt.bar(investor,count)
plt.xlabel('INVESTOR')
plt.ylabel('NO_of_TIME_INVEST')
plt.xticks(rotation=30)
plt.show()
```
