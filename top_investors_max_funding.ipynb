```python
"""In this problem we have to find the top 5 investors who have invested maximum number of times in
different companies which means, if one investor has invested multiple times in one startup, we have to 
count one for that company."""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

"""Function check is used to covert the different number of investors present in a single startup into a list,
 this help us to iterate over different investors easily."""

def function(investor):
    list1=investor.split(",")
    return list1
df=pd.read_csv('downloads/startup_funding.csv')
df.StartupName.fillna('/',inplace=True)
df1=df.copy()

"""We have to replace all the errors present in the name of important startups such as Ola,flipkart,etc."""

df.StartupName.replace("Ola Cabs","Ola",inplace=True)
df.StartupName.replace("Oyo Rooms","Oyo",inplace=True)
df.StartupName.replace("Flipkart.com","Flipkart",inplace=True)
df.StartupName.replace("Olacabs","Ola",inplace=True)
df.StartupName.replace("OYO Rooms","Oyo",inplace=True)
df.StartupName.replace("OyoRooms","Oyo",inplace=True)
df.StartupName.replace("Oyorooms","Oyo",inplace=True)
df.StartupName.replace("Ola Cabs","Ola",inplace=True)
df.StartupName.replace("Paytm Marketplace","Paytm",inplace=True)
d={}

""" We have to fill all Nan entries present in Investors Name """

df.InvestorsName.fillna('/',inplace=True)
df.InvestorsName=df.InvestorsName.apply(function)

""" Now we have to use a dictionary such that it contains the key(InvestorsName) and key values as StartUpName """

for i in df.InvestorsName.index:
    for j in df.InvestorsName[i]:
        if (j!="/") :
            if(df.StartupName[i] not in d.get(j.strip(),'/')):
                d[j.strip()]=(d.get(j.strip(),'/')) +'/'+ (df.StartupName[i])
del d[""]

""" we use np arrays to store the values of keys in names as a list values """

names=np.array(list(d.keys()))
list1=list(d.values())
vlues=np.array([len(i.split("/")) for i in list1])
xaxis=[]
yaxis=[]
for i in range(0,5):
    index=vlues.argmax()
    print(names[index])
    xaxis.append(names[index])
    yaxis.append(vlues[index])
    vlues[index]=0
plt.pie(yaxis,labels=xaxis,autopct="%.2f"'%%')
plt.show()
plt.bar(xaxis,yaxis)
plt.xticks(rotation=10)
plt.show()
```
